```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_course = pd.read_csv('../../employeeEngagement/prep/courses.csv')

df_course_progress = pd.read_csv('../../employeeEngagement/prep/courseProgress.csv')
```

```python
merged_df = pd.merge(
    df_course, df_course_progress,
    left_on=['id'],
    right_on=['courseId'],
    how='inner'
)

grouped_counts = merged_df.groupby(['courseId', 'title']).size().reset_index(name='count')

grouped_counts = grouped_counts.sort_values(by='count', ascending=False).head()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='title', y='count', data=grouped_counts)

# Adding labels and title
plt.xlabel('Course Name')
plt.ylabel('Count of Employees')
plt.title('Employee Count per Course')
plt.xticks(rotation=70)

# Display the plot
plt.show()
```
