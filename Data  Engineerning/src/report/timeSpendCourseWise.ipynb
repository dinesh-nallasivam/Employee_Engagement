```python
import pandas as pd
```

```python

df_course = pd.read_csv('../../employeeEngagement/prep/courses.csv')

df_courseProgres = pd.read_csv('../../employeeEngagement/prep/courseProgress.csv')
```

```python
df_courseProgres = df_courseProgres.loc[df_courseProgres['progress']==100]

df_courseProgres = df_courseProgres.groupby(['courseId'])['timeSpent'].agg(
    max_time_in_hr = 'max',
    min_time_in_hr = 'min'
)
```

```python
def convert_seconds_to_hr_min(seconds):
    hours = seconds // 3600
    minutes = (seconds % 3600) // 60
    return f"{hours} hr {minutes} min"

df_courseProgres['min_time_in_hr'] = df_courseProgres['min_time_in_hr'].apply(convert_seconds_to_hr_min)

df_courseProgres['max_time_in_hr'] = df_courseProgres['max_time_in_hr'].apply(convert_seconds_to_hr_min)
```

```python
final_merge = pd.merge(
    df_course, df_courseProgres,
    left_on=['id'],
    right_on=['courseId'],
    how='inner'
)

final_merge = final_merge.drop(columns=['totalQuiz','totalChapter','startedAt','description','Unnamed: 0'])

display(final_merge)
```
