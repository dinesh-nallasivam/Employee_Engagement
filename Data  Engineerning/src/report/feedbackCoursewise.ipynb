```python
import pandas as pd
```

```python
df_course = pd.read_csv('../../employeeEngagement/prep/courses.csv')

df_feedback = pd.read_csv('../../employeeEngagement/prep/feedbacks.csv')
```

```python
df_feedback = df_feedback.groupby(['courseId'])['response'].mean().reset_index(name='rating').round(2)

merged_df = pd.merge(
    df_course,df_feedback, left_on=['id'], right_on=['courseId'], how='left' 
)

merged_df.drop(columns=['Unnamed: 0','totalQuiz'],inplace=True)

display(merged_df)
```

```python
final_merge = pd.merge(
    df_course, df_feedback,
    left_on=['id'],
    right_on=['courseId'],
    how='inner'
)

final_merge = final_merge.drop(columns=['totalQuiz','totalChapter','startedAt','description','Unnamed: 0'])

display(final_merge)
```
