```python
import pandas as pd
```

```python
df_employee = pd.read_csv('../../employeeEngagement/prep/employees.csv')

df_courseProgres = pd.read_csv('../../employeeEngagement/prep/courseProgress.csv')
```

```python
df_courseProgres = df_courseProgres.drop_duplicates(subset=['employeeId'])
```

```python

final_merge = pd.merge(
    df_employee, df_courseProgres,
    left_on=['id'],
    right_on=['employeeId'],
    how='inner'
)
```

```python

final_merge = final_merge.groupby(['department']).size().reset_index(name="employee_count")

display(final_merge)
```
